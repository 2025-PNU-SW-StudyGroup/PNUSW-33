```python
f = open('./birdclef-2025/train.csv', 'r', encoding='utf-8-sig')
reading = csv.reader(f)
clefFile = []
for line in reading:
    clefFile.append(line)
# clefFile = np.array(clefFile)
print(clefFile[:3])
```

```python
fileLocation = list(zip(*clefFile[1:]))[3]
print(f'{fileLocation[:10]}')
```

```python
from spleeter.separator import Separator
import os

def separate_all(file_list):
    separator = Separator('spleeter:2stems')  # 한 번만 초기화
    i = 0
    for file in file_list:
        try:
            i += 1
            print(f"▶ 시작: {file} - {i}")
            audio_path = os.path.join('birdclef-2025', 'train_audio', file)
            output_path = os.path.join('output', file.split('/')[0])
            os.makedirs(output_path, exist_ok=True)

            separator.separate_to_file(audio_path, output_path)
        except Exception as e:
            print(f"[에러 발생] {file}: {e}")
```

```python
if __name__ == '__main__':
    separate_all(fileLocation)
```

```python
for file in fileLocation:
    path = './output/' + file[:-4]
    os.rename(path + '/accompaniment.wav', path + '/' + file.split('/')[1][:-4]+'.wav')
    print(file.split('/')[1][:-4], '완료')
```
