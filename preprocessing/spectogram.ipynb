```python
import os
os.mkdir('spectrograms')
root = os.listdir('birdclef-2025/train_audio')
for i in root:
    os.mkdir('spectrograms/'+i)
```

```python
import soundfile as sf
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import gc
import matplotlib; matplotlib.use('Agg')

def ogg_to_spectrogram(ogg_path, save_path=None):
    # soundfile로 로드
    y, sr = sf.read(ogg_path)
    if len(y.shape) > 1:
        y = y.mean(axis=1)  # stereo -> mono

    # librosa.stft 사용
    D = librosa.stft(y)
    S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

    fig, ax = plt.subplots(figsize=(10, 6))
    img = librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='hz', cmap='magma', ax=ax)
    fig.colorbar(img, ax=ax, format='%+2.0f dB')
    ax.set(title='Spectrogram (dB)')
    fig.tight_layout()

    if save_path:
        fig.savefig(save_path)
    else:
        plt.show()
    plt.close('all')

    del y, sr, D, S_db, fig, ax, img
    gc.collect()
```

```python
import os
import csv
f = open('./birdclef-2025/train.csv', 'r', encoding='utf-8-sig')
reading = csv.reader(f)
clefFile = []
for line in reading:
    clefFile.append(line)
# clefFile = np.array(clefFile)
print(clefFile[:3])
```

```python
fileLocation = list(zip(*clefFile[1:]))[3]
print(f'{fileLocation[:10]}')
```

```python
i = 'birdclef-2025/train_audio/'
for file in fileLocation:
    o = "spectrograms/"+file[:-3]+"png"
    ogg_to_spectrogram(i+file, o)
    print(f'{file}: Done')
```
